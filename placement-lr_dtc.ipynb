# Encoding by One Hot Encoder
# Encoding is a process of Converting Categorical variable into Numerical Variable

```python
X=df.drop(columns=["status"])  # Independen Feature
y=df.status    # Dependent Feature
print(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#,"degree_t","workex","specialisation"
ohe=OneHotEncoder(sparse=False,dtype=np.int32)
X_new=ohe.fit_transform(X[["hsc_s","degree_t","workex","specialisation"]])
print(X_new.shape)
value_x=np.hstack((X[["gender","ssc_p","hsc_p","degree_p","etest_p","mba_p","salary"]].values,X_new))
print(value_x)
```

# Dividing Dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(value_x,y,test_size=0.2,random_state=2)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

# Creating a model Using Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
ypred_lr=lr.predict(xtest)
```

# Creating a model Using DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(xtrain,ytrain)
```

```python
ypred_dtc=dtc.predict(xtest)
```

# Performace Metric to Check the Performance of the Models

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

# Performance of Logistic Regression Model

```python
confusion_matrix(ytest,ypred_lr)
```

```python
accuracy_score(ytest,ypred_lr)
```

```python
print(classification_report(ytest,ypred_lr))
```

# Performance of DecisionTreeClassifier Model

```python
confusion_matrix(ytest,ypred_dtc)
```

```python
accuracy_score(ytest,ypred_dtc)
```

```python
print(classification_report(ytest,ypred_dtc))
```
